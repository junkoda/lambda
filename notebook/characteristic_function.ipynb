# Characteristic function

``lambda.lib.load_characterstic_funtion`` loads the characteristic function of $u(x)$


$$
\varphi(\lambda) = \left\langle e^{i\lambda u(x)} \right\rangle
$$

- 'lambda' (array): $\lambda$
- 'summary' (dict): mean (`phi`) and the standard error in the mean (`dphi`) of multiple realisations
- 'phi' (array): characteristic function `phi[1, ilambda, irealisation]`
- 'sigma2_v' (float): displacement field rms $\sigma^2_v = \left\langle u(x)^2 \right\rangle#


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import lambdalib


isnp= '010'
d = lambdalib.load_characteristic_function('wizcola', isnp)

lambdas = d['summary']['lambda']
phi = d['summary']['phi']   # mean characteristic function of 14 realisations
sigma2_v = d['summary']['sigma2_v']
gaussian = np.exp(-0.5*sigma2_v*lambdas**2)

plt.figure()
plt.xlim(0, 2)
plt.xlabel('$\\lambda$')
plt.ylabel('$\\varphi$')

plt.plot(lambdas, phi, label='summary')
plt.plot(d['phi'][:, 0, 0], d['phi'][:, 1, 0], ':', label='one realisation')
plt.plot(lambdas, gaussian, '--', label='Gaussian')

plt.legend()
plt.show()
```
